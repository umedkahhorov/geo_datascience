```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
liney = np.arange(1,10,1).reshape(1,-1)
linex = np.arange(1,10,1).reshape(1,-1)
plt.plot(linex,liney,'ro',markersize=6)

pointx = np.arange(1,10,0.1) + np.random.randn(90)
pointx = pointx.reshape(-1,1)
pointy = np.arange(1,10,0.1) + np.random.randint(low=-3, high=3, size=(90,))
pointy = pointy.reshape(-1,1)
plt.plot(pointx,pointy,'bo',markersize=2)


dist = np.sqrt( (liney - pointy)**2 + (linex - pointx)**2)
idx = np.argmin(dist,axis=0)
newx = pointx.squeeze()[idx]
newy = pointy.squeeze()[idx]

plt.scatter(newx,newy,s=60, facecolors='none', edgecolors='r')
```

```python
def search_closest_points(lineX,lineY,pointX,pointY):
    """
    get xy closest matched to np.polyfit(x,y)
    """
    lineX = np.array(lineX).reshape(1,-1)
    lineY = np.array(lineY).reshape(1,-1)
    pointX = np.array(pointX).reshape(-1,1)
    pointY = np.array(pointY).reshape(-1,1)
    
    dist = np.sqrt( (lineY - pointY)**2 + (lineX - pointX)**2)
    idx = np.argmin(dist,axis=0)

    newx = pointX.squeeze()[idx]
    newy = pointY.squeeze()[idx]
    
    return newx,newy
```

```python
df = pd.read_csv('test.csv') # xy data
df = df.round(2)
x = df['SRC_X'] 
y = df['SRC_Y']
print (df.shape)
# fit a line
a,b = np.polyfit(x,y,deg=1)
dx=200
x1 = np.arange(x.min(),x.max()+dx,dx) 
y1 = a*x1+b
# get closest point to fitted line
x2,y2= search_closest_points(x1,y1,x,y)
xy = pd.DataFrame({'SRC_X':x2,'SRC_Y':y2})
xy = xy.round(2)
(print (xy.shape))
xy['Picked'] = 1
# return dataframe with special index 
df = df.merge(xy,how='outer').fillna(0)
print (df.shape)
```
